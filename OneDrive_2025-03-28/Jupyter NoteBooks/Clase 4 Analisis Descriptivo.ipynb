# Análisis del Dataset Iris

El dataset **Iris** es un conjunto de datos ampliamente utilizado en la ciencia de datos y el aprendizaje automático.
Fue introducido por el estadístico y biólogo Ronald Fisher en 1936 y contiene información sobre 150 muestras de flores de iris de tres especies diferentes.

## Descripción de las Columnas

El dataset tiene cinco columnas:

1. **sepal_length**: Longitud del sépalo en centímetros.
2. **sepal_width**: Ancho del sépalo en centímetros.
3. **petal_length**: Longitud del pétalo en centímetros.
4. **petal_width**: Ancho del pétalo en centímetros.
5. **species**: Especie de la flor (setosa, versicolor, virginica).

https://www.kaggle.com/datasets/uciml/iris

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Carga del conjunto de datos Iris desde seaborn
df = sns.load_dataset('iris')
```

```python
# 1. Análisis de tipos de datos
print("Tipos de datos en cada columna:")
print(df.dtypes)
```

```python
# 2. Medidas estadísticas generales (solo para columnas numéricas)
print("Estadísticas descriptivas generales:")
df.describe()
```

```python
# 3. Análisis de valores nulos
print("Conteo de valores nulos por columna:")
df.isnull().sum()
```

```python
# 4. Análisis de la media y mediana de las características numéricas
print("Media de las características numéricas:")
print(df.select_dtypes(include=[np.number]).mean())

print("\nMediana de las características numéricas:")
print(df.select_dtypes(include=[np.number]).median())
```

```python
# 5. Análisis de la varianza y desviación estándar
print("Varianza de las características numéricas:")
print(df.select_dtypes(include=[np.number]).var())

print("\nDesviación estándar de las características numéricas:")
print(df.select_dtypes(include=[np.number]).std())
```

```python
# 6.1 Histograma de la longitud del sépalo
plt.figure(figsize=(10, 6))
sns.histplot(df['sepal_length'], kde=True, color='blue', bins=20)
plt.title('Histograma de la Longitud del Sépalo')
plt.xlabel('Longitud del Sépalo (cm)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
## Group by species
df_grouped = df.groupby('species').count()
df_grouped
```

```python
# 6.2 Gráfico de barras de la especie
plt.figure(figsize=(10, 6))
sns.countplot(x='species', data=df, hue='species', palette='Set2', legend=False)
plt.title('Conteo de Especies')
plt.xlabel('Especie')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# 6.3 Gráfico de dispersión entre la longitud y el ancho del pétalo
plt.figure(figsize=(10, 6))
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=df, palette='Set1')
plt.title('Relación entre Longitud y Ancho del Pétalo')
plt.xlabel('Longitud del Pétalo (cm)')
plt.ylabel('Ancho del Pétalo (cm)')
plt.legend(title='Especie')
plt.show()
```

```python
# 6.4 Stripplot de la longitud del sépalo por especie
plt.figure(figsize=(10, 6))
sns.stripplot(x='species', y='sepal_length', data=df, hue='species', palette='Set1')
plt.title('Distribución de la Longitud del Sépalo por Especie (Stripplot)')
plt.xlabel('Especie')
plt.ylabel('Longitud del Sépalo (cm)')
plt.legend(title='Especie')
plt.show()
```

# Diagrama de cajas (BoxPlot) O diagrama de cajas y bigotes (Box and Whisker)



```python
# 6.5 Boxplot de la longitud del sépalo por especie
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='sepal_length', data=df, hue='species', palette='Set3', dodge=False)
plt.title('Boxplot de la Longitud del Sépalo por Especie')
plt.xlabel('Especie')
plt.ylabel('Longitud del Sépalo (cm)')
plt.show()
```

### Valor Atipico o Outlier
es un punto que se encuentra significativamente alejado de los demas, y se representan como puntos fuera de los bigotes

```python
# 6.6 Boxplot y Stripplot combinados
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='sepal_length', data=df, palette='Set1', dodge=False)
sns.stripplot(x='species', y='sepal_length', data=df, hue='species', palette='Set1')
plt.title('Boxplot y Stripplot de la Longitud del Sépalo por Especie')
plt.xlabel('Especie')
plt.ylabel('Longitud del Sépalo (cm)')
plt.show()
```

```python
# 6.7 Gráfico de violin de la longitud del sépalo
plt.figure(figsize=(10, 6))
sns.violinplot(x='species', y='sepal_length', data=df, palette='pastel')
plt.title('Distribución de la Longitud del Sépalo por Especie')
plt.xlabel('Especie')
plt.ylabel('Longitud del Sépalo (cm)')
plt.show()
```

```python
# 6.8 Pairplot para visualizar las relaciones entre todas las características
sns.pairplot(df, hue='species', palette='Set1')
plt.suptitle('Pairplot de las Características del Iris', y=1.02)
plt.show()
```

```python
# 6.9 Mapa de calor de la correlación entre variables numéricas
plt.figure(figsize=(8, 6))
sns.heatmap(df.select_dtypes(include=[np.number]).corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Mapa de Calor de Correlación entre Variables')
plt.show()
```

```python
df_corr = pd.DataFrame(np.random.randn(100, 2), columns=['x', 'y'])
df_corr['y'] = df_corr['x']
plt.figure(figsize=(10, 6))
sns.scatterplot(x='x', y='y', data=df_corr)
plt.title('Relación con Correlación de 1')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```

```python
##Create a plot with -1 correlation

df_corr['y'] = -df_corr['x']
plt.figure(figsize=(10, 6))
sns.scatterplot(x='x', y='y', data=df_corr)
plt.title('Relación con Correlación de -1')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```
